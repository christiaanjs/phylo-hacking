# Sparse variational learning on categorical latent variables

```python
import numpy as np
import autograd.numpy as anp
import autograd
from autograd.extend import primitive, defvjp
import matplotlib.pyplot as plt
%matplotlib inline
```

## Sparsemax function

```python
test = np.array([[1, 4, 2],
                 [-2, -10, 3]]).astype(float)

def sort_rows(z):
    perm = anp.argsort(z, axis=1)
    rows = anp.arange(z.shape[0], dtype=int).reshape(-1, 1)
    return z[rows, perm]

def sparsemax_rows(z):
    ncols = z.shape[1]
    u = sort_rows(z)[:, ::-1]
    cssv = u.cumsum(axis=1) - 1
    ind = anp.arange(ncols) + 1
    cond = u - cssv / ind > 0
    argmin = anp.argmin(cond, axis=1)
    rho = np.where(argmin > 0, argmin, ncols)
    theta = cssv[anp.arange(z.shape[0]) , rho - 1]/rho
    return anp.clip(z - theta.reshape(-1, 1), 0, None)

sparsemax_rows(test), autograd.jacobian(sparsemax_rows)(test)
```

## Experimenting with penalties

```python
from scipy.special import entr

smaller_p = 0.05
small_p = 0.1

def transform(p1, p2):
    other = (1-p1-p2)/2
    return np.stack([p1, p2, other, other], axis=-1).clip(min=0) # Clip to fix floating point issues

true_posterior = transform(smaller_p, small_p)

def log_p(ps):
    return (ps * np.log(true_posterior)).sum(axis=-1)

def log_q(ps):
    return -(entr(ps)).sum(axis=-1)

def kl(ps):
    return log_q(ps) - log_p(ps)

epsilon = 1e-2

def penalty(ps, reg): #Pseudo-dirichlet
    return -(reg - 1)*anp.log(epsilon + ps).sum(axis=-1)

p1s = np.arange(0, 1.001, 0.01)
p2s = np.arange(0, 1.001, 0.01)
P1s, P2s = np.meshgrid(p1s, p2s)
valid = P1s + P2s <= 1.0

ps = transform(P1s, P2s)
kls = np.where(valid, kl(ps), np.inf)

fig, axs = plt.subplots(ncols=3, nrows=3, figsize=(20, 20))

regs = [0.99, 0.95, 0.9]

axs[0, 0].set_title('KL divergence')
axs[0, 1].set_title('Penalty')
axs[0, 2].set_title('Penalized objective')

for i in range(3):
    axs[i, 0].set_ylabel('Alpha: ' + str(regs[i]))
    
    penalties = np.where(valid, penalty(ps, regs[i]), np.inf)
    axs[i, 0].contour(P1s, P2s, kls, 30)
    axs[i, 1].contour(P1s, P2s, penalties, 30)
    obj = kls + penalties
    axs[i, 2].contour(P1s, P2s, obj, 30)
    
    argmin_ij = np.unravel_index(np.argmin(obj), obj.shape)
    argmin = (P1s[argmin_ij], P2s[argmin_ij])
    
    for j in range(3):
        axs[i, j].scatter([argmin[0]], [argmin[1]], marker='x')
        axs[i, j].annotate('Min: ' + str(ps[argmin_ij]), xy=argmin)
        axs[i, j].scatter([smaller_p], [small_p])
```

```python
from autograd.scipy.special import gammaln

epsilon = 1e-3
log_likelihoods = norm.logpdf(y.reshape(-1, 1), loc=mus.reshape(1, -1), scale=sd)

def is_sparse(ps, **kwargs):
    return anp.all((ps == 0) | (ps == 1), axis=-1, **kwargs)

def penalty(ps, reg=0.5): # Reg in (0, 1.0) - lower is stronger
    return np.where(is_sparse(ps, keepdims=True), 0.0, anp.power(anp.power(ps, reg).sum(axis=-1), 1/reg)).sum()

def logq_prior(ps, reg): #Pseudo-dirichlet
    return (reg - 1)*anp.log(epsilon + ps).sum()

def logq_likelihood_e(ps):
    nonzero_ps = ps[ps > 0]
    return anp.sum(nonzero_ps * anp.log(nonzero_ps))

def logp_e(ps):
    nonzero = ps > 0
    return (ps[nonzero] * np.log(true_posterior[nonzero])).sum() #Is this correct marginalisation?

def kl(ps):
    return logq_likelihood_e(ps) - logp_e(ps)

def kl_reg(ps, reg=0.5):
    return logq_likelihood_e(ps) - logp_e(ps) - logq_prior(ps, reg=reg)

kl(true_posterior), kl_reg(true_posterior), autograd.grad(kl_reg)(true_posterior, reg=0.7)
```

```python
from tqdm import tqdm_notebook as tqdm

# https://github.com/bamos/opt.py/blob/master/bamos_opt/pgd.py
def pgd(x0, f, g, proj, eps, maxit, lam=None, acc=True,
          doLinesearch=True, callback=None, **kwargs):
    """TODO."""
    results = {
        'feval': 0,
        'geval': 0
    }

    x = proj(np.copy(x0))
    prev_x = x

    results['bestF'] = None
    results['bestX'] = np.copy(x)

    for k in tqdm(range(maxit)):
        omega = (k - 1.0) / (k + 2.0)
        d = x - prev_x
        y = x if acc else x + omega * d
        if k > 1 and np.linalg.norm(d) < eps:
            break
        f_y = f(y, **kwargs)
        results['feval'] += 1
        g_y = g(y, **kwargs)
        results['geval'] += 1
        prev_x = x
        if doLinesearch:
            lam = 1
            beta = 0.5
            while True:
                x = proj(y - lam * g_y)
                dxy = x - y
                f_x = f(x, **kwargs)
                results['feval'] += 1
                if f_x <= f_y + (g_y * dxy).sum() + np.linalg.norm(dxy)**2 / (2 * lam):
                    break
                lam = beta * lam
        else:
            x = proj(y - lam * g_y)
            f_x = f(x, **kwargs)
            results['feval'] += 1

        if results['bestF'] is None or f_x < results['bestF']:
            results['bestF'] = f_x
            results['bestX'][:] = x

        if callback:
            callback(k, results['bestF'], y, g_y, lam)

    return results

max_iter = 10000
results = np.zeros((max_iter,) + true_posterior.shape)

def callback(i, f, x, grad, lam):
    results[i] = x

opt_res = pgd(np.ones_like(true_posterior)/3,
    kl,
    autograd.grad(kl),
    sparsemax_rows,
    1e-7,
    max_iter,
    callback=callback)
opt_res
```

## Stochastic variational inference

```python
rows = anp.arange(n)

def logp(assignments):
    return anp.sum(norm.logpdf(y, loc=mus[assignments], scale=sd) + anp.log(p)[assignments])

def logq(assignments, lambd):
    assignment_ps = sparsemax_rows(lambd)
    return anp.sum(anp.log(assignment_ps[rows, assignments]))

def sample_row_p(x, p):
    u = np.random.uniform(size=p.shape[0])
    pf = p.cumsum(axis=1)
    return np.argmax(u.reshape(-1, 1) < pf, axis=1)

def sample_q(lambd):
    p = sparsemax_rows(lambd)
    return sample_row_p(x, p)

# Approximate expectation w.r.t. dens(z; theta) of func(z) using samples[i = 1:s, ...] ~ dens(z; theta)

def score_function_estimator(func, sample, logdens_grad, theta):
    return func(sample)*logdens_grad(sample, theta)

entropy_grad = autograd.grad(entropy_q)
logq_grad = autograd.grad(logq, argnum=1)

def kl_grad(lambd, i):
    assignments = sample_q(lambd)
    return autograd.grad(entropy_q)(lambd) - score_function_estimator(logp, assignments, logq_grad, lambd)

kl_grad(np.random.normal(size=(n, 3)), 1)
```

```python
lambd_init = np.zeros((n, len(p)))
kl_grad(lambd_init, 0)
```

```python
def get_callback(every):
    def callback(lambd, i, grad):
        if i % every == 0:
            assignments_ps = sparsemax_rows(lambd)
            print('Iteration {}'.format(i))
            print(assignments_ps)
    return callback
```

```python
from autograd.misc.optimizers import rmsprop
        
res = rmsprop(kl_grad, lambd_init, callback=get_callback(1000), step_size=1e-3, num_iters=100000)
sparsemax_rows(res)
```
